```python
# Dataset de exemplo
data = vinhoBranco["quality"]

# Calcule a mÃ©dia e o desvio padrÃ£o dos dados
mean = np.mean(data)
std = np.std(data)

# Defina o limite superior e inferior usando a regra dos trÃªs desvios padrÃ£o
upper_limit = mean + 5 * std
lower_limit = mean - 5 * std

# Encontre os outliers
outliers = [x for x in data if x > upper_limit or x < lower_limit]

# Imprima os resultados
print("MÃ©dia:", mean)
print("Maior: ", data.max())
print("Menor: ", data.min())
print("Desvio padrÃ£o:", std)
print("Limite superior:", upper_limit)
print("Limite inferior:", lower_limit)
print("Outliers:", len(outliers))
plt.plot(data)

# Adiciona um tÃ­tulo ao grÃ¡fico
plt.title('Meu grÃ¡fico')

# Adiciona rÃ³tulos aos eixos x e y
plt.xlabel('Tempo')
plt.ylabel('Valor')

# Exibe o grÃ¡fico
plt.show()
```
