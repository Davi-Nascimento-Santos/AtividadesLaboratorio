```python
import matplotlib.pyplot as plt
import numpy as np
from Kmeans import *

#lista = []

# Cria um array com dados de x e y
array = [[1, 4], [2, 6], [1, 9], [2, 8], [3, 4]]
centroides = [[3, 2], [4, 6]]
clusters = mostClose(array, centroides)
x1 = []
y1 = []
x2 = []
y2 = []
for i in clusters[0]:
    x1.append(i[0])
    y1.append(i[1])
for i in clusters[1]:
    x2.append(i[0])
    y2.append(i[1])
#print(x)
#print(y)
# Cria um scatterplot com x e y
plt.scatter(x1, y1)
plt.scatter(x2, y2)
# Define os rótulos dos eixos
plt.xlabel('X')
plt.ylabel('Y')

# Define o título do gráfico
plt.title('Exemplo de Scatterplot')

# Mostra o gráfico
plt.show()
```
